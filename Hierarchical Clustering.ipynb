```python
# pandas 사용하기
import numpy as np # numpy 도 함께 import
import pandas as pd
from scipy.cluster.hierarchy import linkage
from scipy.spatial.distance import pdist, squareform
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
import time
```

```python
row_dist = pd.read_csv('data.csv')
```

```python
row_dist
```

```python
distanceList = ['complete', 'single', 'average', 'weighted', 'centroid', 'median', 'ward']
```

```python
for dis in distanceList:
    start = time.time()
    # linkage함수에서 초기 데이터 배열을 전달하고, euclidean 지표를 매개변수로 사용
    row_clusters = linkage(row_dist.values, metric = 'euclidean', method = dis)
    row_dendr = sch.dendrogram(row_clusters)
    plt.tight_layout()
    plt.ylabel(dis + ' distance')
    plt.show()
    end = time.time()
    print(f"{end - start:.5f} sec")
```

```python
start = time.time()
# linkage함수에서 초기 데이터 배열을 전달하고, euclidean 지표를 매개변수로 사용
row_clusters = linkage(row_dist.values, metric = 'euclidean', method = 'ward')
row_dendr = sch.dendrogram(row_clusters)
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
end = time.time()
print(f"{end - start:.5f} sec")
```
